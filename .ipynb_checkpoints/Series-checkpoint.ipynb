## Series 

```python
ice_cream = ["Chocolate", "Vanilla", "Strawberry", "Rum Raisin"]
pd.Series(ice_cream)
```

```python
numbers = [100,200,300,400]
pd.Series(numbers)
```

```python
#dtype is "object" for string list
```

```python
webster = {"Horse": "an animal",
          "Apple": "a fruit",
          "blue": "a color"}

pd.Series(webster)
```

```python
s = pd.Series(ice_cream)

#some attributes

print s.values
print s.index
print s.dtype
```

```python
prices = [2.99, 4.45, 1.36]
s = pd.Series(prices)
s
```

```python
#some methods

s.sum()
```

```python
s.product()
```

```python
s.mean()
```

```python
fruits = ["apple", "banana", "orange","mango"]
days = ["monday","tuesday","wednesday","thursday"]

pd.Series(fruits, days)                           #shift + tab (+ tab + tab) inside parenthesis
```

```python
days = ["monday","tuesday","wednesday","monday"]
pd.Series(fruits, days)  #this is difference between lists and series. Here, same index is allowed twice!
```

```python
def classify_perf(number):
    if number < 300:
        return "OK"
    elif number >= 300 and number < 650:
        return "Satisfactory"
    else:
        return "incredible!"
```

```python
google.apply(classify_perf).tail()
```
