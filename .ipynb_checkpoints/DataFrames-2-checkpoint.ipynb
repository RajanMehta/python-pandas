```python
import pandas as pd
```

```python
df = pd.read_csv("data/employees.csv", parse_dates=["Start Date", "Last Login Time"])  #using parse_dates instead of lines 2 & 3.

# df["Start Date"] = pd.to_datetime(df["Start Date"])
# df["Last Login Time"] = pd.to_datetime(df["Last Login Time"])

df["Senior Management"] = df["Senior Management"].astype("bool")
df["Gender"] = df["Gender"].astype("category")
df.head(3)
```

```python
df.sort_values("First Name", inplace=True)

df["First Name"].duplicated()    
# keep = "first" => first occurence "false" i.e not duplicate. After that, all occurences "true" i.e. duplicates

df["First Name"].duplicated(keep="last")
# keep = "last" => last occurence "false"

df["First Name"].duplicated(keep=False)
# keep = "fasle" => mark all multiple entries as "true"

df[~df["First Name"].duplicated(keep=False)]
# rows that only occur once
```
