### Extract rows from multiIndex

```python
bigmac = pd.read_csv("data/bigmac.csv", parse_dates = ["Date"], index_col=["Date", "Country"])
bigmac.sort_index(inplace=True)
bigmac.head(3)
```

```python
bigmac.loc[("2010-01-01", "Brazil")]
```

### Transpose

```python
bigmac_tr = bigmac.transpose()
bigmac_tr
```

```python
bigmac_tr.loc["Price in US Dollars", ("2010-01-01","Brazil")]
```

### Stack Method

```python
world = pd.read_csv("data/worldstats.csv", index_col = ["country", "year"])
world.head(3)
```

```python
world.stack()    # will convert to a Series with 3-level index
```

```python
world.unstack(1)
```

### Pivot Method

```python
sales = pd.read_csv("data/salesmen.csv", parse_dates = ["Date"])
sales.head(3)
```

```python
sales.pivot(index = "Date", columns = "Salesman", values = "Revenue").head()
```

### pivot_table method

```python
food = pd.read_csv("data/foods.csv")
food.head()
```

```python
food.pivot_table(index="Gender", values="Spend", aggfunc="sum")
```

```python
food.pivot_table(index=["Gender","Item"], values="Spend", aggfunc="sum")
```

```python
food.pivot_table(index=["Gender","Item"], columns="City", values="Spend", aggfunc="sum")
```

```python
food.pivot_table(index=["Gender","Item"], columns=["Frequency","City"], values="Spend", aggfunc="sum")
```
